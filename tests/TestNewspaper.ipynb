This notebook is a testing sandbox for the newspaper python package

```python
import newspaper
from IPython.display import display, HTML
```

```python
ArsTechnica = newspaper.build('http://arstechnica.com/', memoize_articles=False)
AL_paper = newspaper.build('http://www.avionslegendaires.net/', language='fr', memoize_articles=False)
polar_blog = newspaper.build('https://www.polar.com/blog/', memoize_articles=False)
theverge = newspaper.build('http://www.theverge.com/', memoize_articles=False)

print("Retrived %i articles from arstechnica" % ArsTechnica.size())
print("Retrived %i articles from avionslegendaires" % AL_paper.size())
print("Retrived %i articles from polar blog" % polar_blog.size())
print("Retrived %i articles from the verge" % theverge.size())

articles = ArsTechnica.articles + AL_paper.articles + polar_blog.articles + theverge.articles

for article in articles:
    print(article.url)
    
```

```python
for category in ArsTechnica.category_urls():
    print(category)
    
for category in theverge.category_urls():
    print(category)
    
```

```python
for feed_url in ArsTechnica.feed_urls():
    print(feed_url)
for feed_url in AL_paper.feed_urls():
    print(feed_url)
for feed_url in polar_blog.feed_urls():
    print(feed_url)
for feed_url in theverge.feed_urls():
    print(feed_url)
```
