This notebook is a testing sandbox for the newspaper python package

```python
import newspaper
from IPython.display import display, HTML
import pandas as pd
```

```python
ArsTechnica = newspaper.build('http://arstechnica.com/', memoize_articles=False)
AL_paper = newspaper.build('http://www.avionslegendaires.net/', language='fr', memoize_articles=False)
polar_blog = newspaper.build('https://www.polar.com/blog/', memoize_articles=False)
theverge = newspaper.build('http://www.theverge.com/', memoize_articles=False)


print("Retrived %i articles from arstechnica" % ArsTechnica.size())
print("Retrived %i articles from avionslegendaires" % AL_paper.size())
print("Retrived %i articles from polar blog" % polar_blog.size())
print("Retrived %i articles from the verge" % theverge.size())

articles = ArsTechnica.articles + AL_paper.articles + polar_blog.articles + theverge.articles

for article in articles:
    print(article.url)
```

```python
for category in ArsTechnica.category_urls():
    print(category)
    
for category in theverge.category_urls():
    print(category)
```

```python
titles = []
texts = []
top_images = []
authors = []
movies = []
images = []
summaries = []
keywords = []
urls = []
i=1
```

```python
for article in articles:
    try:
        article.download()
        article.parse()
        article.nlp()
        titles.append(article.title)
        texts.append(article.text)
        top_images.append(article.top_image)
        authors.append(article.authors)
        movies.append(article.movies)
        images.append(article.images)
        summaries.append(article.summary)
        keywords.append(article.keywords)
        urls.append(article.url)
        print('Article '+str(i)+' out of '+str(len(articles)))
    except:
        print('Erreur Article '+str(i)+' out of '+str(len(articles))+" : "+article.title + ' '+article.url)
    i+=1
```

```python
ArticleStore = pd.DataFrame({'titre':titles, 'texte':texts, 'image':top_images, "auteurs":authors, "vido":movies, "images":images, "rsum": summaries, "mots cl":keywords, "url":urls})
```

```python
ArticleStore.to_pickle('ArticleStore.pkl')
```
