```python
import newspaper
from newspaper import news_pool
```

```python
sourcesURL = [
    ["https://www.aerospatium.info/", "fr"],
    ["http://www.avionslegendaires.net/", "fr"],
    ["http://www.portail-aviation.com/","fr"],
    ["http://supersonique.blogs.challenges.fr/","fr"],
    ["https://www.technologyreview.com/","en"],
    ["https://sciencetonnante.wordpress.com/","fr"],
    ["http://www.internetactu.net/","fr"],
    ["https://krebsonsecurity.com/","en"],
    ["https://www.schneier.com/","en"],
    ["http://thehackernews.com/","en"],
    ["https://blog.ethereum.org/","en"],
    ["http://www.coindesk.com/","en"],
    ["https://www.synology.com/fr-fr/company/news/","fr"],
    ["https://www.cachem.fr/","fr"],
    ["https://www.plex.tv/blog/","en"]
]
papers = []
```

```python
for url in sourcesURL:
    print('Building newspaper for '+url[0])
    paper = newspaper.build(url[0], language=url[1], memoize_articles=False)
    papers.append(paper)

print("Downloading every article")
news_pool.set(papers, threads_per_source=2)
news_pool.join()
```

```python
for paper in papers:
    for article in paper.articles:
        print(article.title)
```
